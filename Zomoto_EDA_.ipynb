```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato=pd.read_csv("E:\Python_ML\ML_Project\Zomato\EDA\zomato.csv")
zomato.head(6)
```

```python
zomato.info()
```

```python
zomato.shape
```

```python
#Deleting Unnecessary columns
zomato=zomato.drop(['url','dish_liked'],axis=1)
```

```python
#looking at dataset after transformation
zomato.head()
```

```python
#Removing the Duplicates
zomato.duplicated().sum()
```

```python
zomato.dropna(how='any',inplace=True)
zomato.info()
```

```python
#Reading column names
zomato.columns
```

```python
#Changing the column names
zomato=zomato.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
zomato.columns
```

```python
zomato['cost']=zomato['cost'].str.replace(',','').astype(np.float) #changing the cost object to string
zomato.info()
```

```python
#Reading unique values from the Rate column
zomato['rate'].unique()
```

```python
#Removing '/5' from Rate
zomato = zomato.loc[zomato.rate!='NEW']
zomato=zomato.loc[zomato.rate!='-'].reset_index(drop=True)
remove_slash=lambda x:x.replace('/5','') if type(x)==np.str else x
zomato.rate=zomato.rate.apply(remove_slash).str.strip().astype('float')
zomato['rate'].head()
```

```python
#Adjust the column names
zomato.name = zomato.name.apply(lambda x:x.title())
#zomato[['online_order','book_table']].replace(('Yes','No'),(True, False),inplace=True)
zomato.online_order.replace(('Yes','No'),(True, False),inplace=True)
zomato.head() # looking at the dataset after transformation
```

```python
zomato.book_table.replace(('Yes','No'),(True, False),inplace=True)
zomato.head()
```

```python
zomato.cost.unique()
```

```python
#Encode the input variable
def Encoded(zomato):
    for i in zomato.columns[~zomato.columns.isin(['rate','cost','votes'])]:
        zomato[i]=zomato[i].factorize()[0]
    return zomato

zomato_en=Encoded(zomato.copy())
zomato_en.head()
```

```python
#Get Correleation between iffrence variable
corr=zomato_en.corr(method='spearman')
plt.figure(figsize=(15,8))
sns.heatmap(corr,annot=True)
zomato_en.columns
```

 Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

Decision Tree regression

```python
from sklearn.tree import DecisionTreeRegressor
dec=DecisionTreeRegressor()
dec.fit(X_train,y_train)
y_pred=dec.predict(X_test)
r2_score(y_test,y_pred)
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
res=RandomForestRegressor()
res.fit(X_train,y_train)
pred=res.predict(X_test)
r2_score(y_test,pred)
```
